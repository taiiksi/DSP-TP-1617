```python
def LowPassfilter_Butterworth (Cutoff, Fs, order):
    f = 1/2 * Fs                # Nyquist theorem: Fs = 2 * f
    Cutoff_Normalized = Cutoff/f
    b,a = signal.butter(order, Cutoff_Normalized, 'low', analog=False)  #a are the denominator coefficients.  
                                                                        #b are the numerator coefficients. 
    w, h = signal.freqz(b,a)    # w = 2pif, h = frequancy response in dB if 20 * np.log10(abs(h))
 
    plt.plot(0.5*Fs*w/np.pi, np.abs(h), 'b')  # w = 2pif with f = 0 to 1. The real frequency is Fs. 
    plt.ylabel('Magnitude')
    plt.xlabel('Frequency [Hz]')
    plt.title('low pass filter_Butterworth')
    
    
    
def HighPassfilter_Butterworth (Cutoff, Fs, order):
    f = 1/2 * Fs                # Nyquist theorem: Fs = 2 * f
    Cutoff_Normalized = Cutoff/f
    b,a = signal.butter(order, Cutoff_Normalized, 'high', analog=False)  #a are the denominator coefficients.  
                                                                        #b are the numerator coefficients. 
    w, h = signal.freqz(b,a)    # w = 2pif, h = frequancy response in dB if 20 * np.log10(abs(h)
    
 
    plt.plot(0.5*Fs*w/np.pi, np.abs(h), 'r')  # w = 2pif with f = 0 to 1. The real frequency is Fs. 
    plt.ylabel('Magnitude')
    plt.xlabel('Frequency [Hz]')
    plt.title('high pass filter_Butterworth')
    
    
    
def BandPassfilter_Butterworth (Cutoff, Fs, order):
    f = 1/2 * Fs                # Nyquist theorem: Fs = 2 * f
    Cutoff_Normalized =  [None] * 2
    Cutoff_Normalized[0] = Cutoff[0]/f
    Cutoff_Normalized[1] = Cutoff[1]/f
    b,a = signal.butter(order, Cutoff_Normalized, 'bandpass', analog=False)  #a are the denominator coefficients.  
                                                                        #b are the numerator coefficients. 
    w, h = signal.freqz(b,a)    # w = 2pif, h = frequancy response in dB if 20 * np.log10(abs(h))
    
    plt.plot(0.5*Fs*w/np.pi, np.abs(h), 'g')  # w = 2pif with f = 0 to 1. The real frequency is Fs. 
    plt.ylabel('Magnitude')
    plt.xlabel('Frequency [Hz]')
    plt.title('bandpass filter_Butterworth')
    
    
    
    
    
    
def BandStopfilter_Butterworth (Cutoff, Fs, order):
    f = 1/2 * Fs                # Nyquist theorem: Fs = 2 * f
    Cutoff_Normalized =  [None] * 2
    Cutoff_Normalized[0] = Cutoff[0]/f
    Cutoff_Normalized[1] = Cutoff[1]/f
    b,a = signal.butter(order, Cutoff_Normalized, 'bandstop', analog=False)  #a are the denominator coefficients.  
                                                                        #b are the numerator coefficients. 
    w, h = signal.freqz(b,a)    # w = 2pif, h = frequancy response in dB if 20 * np.log10(abs(h))
    
   
    plt.plot(0.5*Fs*w/np.pi, np.abs(h), 'k')  # w = 2pif with f = 0 to 1. The real frequency is Fs. 
    plt.ylabel('Magnitude')
    plt.xlabel('Frequency [Hz]')
    plt.title('bandstop filter_Butterworth')
    
    
    
    
def LowPassfilter_Chebychev1(Cutoff, Fs, order, maximum_ripple_allowed):
    f = 1/2 * Fs                # Nyquist theorem: Fs = 2 * f
    Cutoff_Normalized = Cutoff/f
    b,a = signal.cheby1(order, maximum_ripple_allowed, Cutoff_Normalized, 'low', analog=False)  #a are the denominator coefficients.  
                                                                                                 #b are the numerator coefficients. 
    w, h = signal.freqz(b,a)    # w = 2pif, h = frequancy response in dB if 20 * np.log10(abs(h))

    
    plt.plot(0.5*Fs*w/np.pi, np.abs(h), 'b')  # w = 2pif with f = 0 to 1. The real frequency is Fs. 
    plt.ylabel('Magnitude')
    plt.xlabel('Frequency [Hz]')
    plt.title('low pass filter_Chebychev1')
    
    
    

    
    
def HighPassfilter_Chebychev1(Cutoff, Fs, order, maximum_ripple_allowed):
    f = 1/2 * Fs                # Nyquist theorem: Fs = 2 * f
    Cutoff_Normalized = Cutoff/f
    b,a = signal.cheby1(order, maximum_ripple_allowed, Cutoff_Normalized, 'high', analog=False)  #a are the denominator coefficients.  
                                                                                                 #b are the numerator coefficients. 
    w, h = signal.freqz(b,a)    # w = 2pif, h = frequancy response in dB if 20 * np.log10(abs(h))
    

    plt.plot(0.5*Fs*w/np.pi, np.abs(h), 'r')  # w = 2pif with f = 0 to 1. The real frequency is Fs. 
    plt.ylabel('Magnitude')
    plt.xlabel('Frequency [Hz]')
    plt.title('high pass filter_Chebychev1')
    
    
    
    

def BandPassfilter_Chebychev1(Cutoff, Fs, order, maximum_ripple_allowed):
    f = 1/2 * Fs                # Nyquist theorem: Fs = 2 * f
    Cutoff_Normalized =  [None] * 2
    Cutoff_Normalized[0] = Cutoff[0]/f
    Cutoff_Normalized[1] = Cutoff[1]/f
    b,a = signal.cheby1(order, maximum_ripple_allowed, Cutoff_Normalized, 'bandpass', analog=False)  #a are the denominator coefficients.  
                                                                                                     #b are the numerator coefficients. 
    w, h = signal.freqz(b,a)    # w = 2pif, h = frequancy response in dB if 20 * np.log10(abs(h))
    
    plt.plot(0.5*Fs*w/np.pi, np.abs(h), 'g')  # w = 2pif with f = 0 to 1. The real frequency is Fs. 
    plt.ylabel('Magnitude')
    plt.xlabel('Frequency [Hz]')
    plt.title('bandpass filter_Chebychev1')
    
    
    
def BandStopfilter_Chebychev1(Cutoff, Fs, order, maximum_ripple_allowed):
    f = 1/2 * Fs                # Nyquist theorem: Fs = 2 * f
    Cutoff_Normalized =  [None] * 2
    Cutoff_Normalized[0] = Cutoff[0]/f
    Cutoff_Normalized[1] = Cutoff[1]/f
    b,a = signal.cheby1(order, maximum_ripple_allowed, Cutoff_Normalized, 'bandstop', analog=False)  #a are the denominator coefficients.  
                                                                                                     #b are the numerator coefficients. 
    w, h = signal.freqz(b,a)    # w = 2pif, h = frequancy response in dB if 20 * np.log10(abs(h))
   
    plt.plot(0.5*Fs*w/np.pi, np.abs(h), 'k')  # w = 2pif with f = 0 to 1. The real frequency is Fs. 
    plt.ylabel('Magnitude')
    plt.xlabel('Frequency [Hz]')
    plt.title('bandstop filter_Chebychev1')
    
    
    

    
    
    
```

```python

def LowPassfilter_Elliptic(Cutoff, Fs, order, maximum_ripple_allowed, minimum_attenuation): 
    #maximum_ripple_allowed is for the pass band.
    #minimum_attenuation is for the stop band.
    f = 1/2 * Fs                # Nyquist theorem: Fs = 2 * f
    Cutoff_Normalized = Cutoff/f
    b,a = signal.ellip(order, maximum_ripple_allowed, minimum_attenuation, Cutoff_Normalized, 'low', analog=False)  #a are the denominator coefficients.  
                                                                                                 #b are the numerator coefficients. 
    w, h = signal.freqz(b,a)    # w = 2pif, h = frequancy response in dB if 20 * np.log10(abs(h))

    
    plt.plot(0.5*Fs*w/np.pi, np.abs(h), 'b')  # w = 2pif with f = 0 to 1. The real frequency is Fs. 
    plt.ylabel('Magnitude')
    plt.xlabel('Frequency [Hz]')
    plt.title('low pass LowPassfilter_Elliptic')
    
    

def LowPassfilter_Chebychev2(Cutoff, Fs, order, minimum_attenuation):
    #minimum_attenuation is for the stop band.
    f = 1/2 * Fs                # Nyquist theorem: Fs = 2 * f
    Cutoff_Normalized = Cutoff/f
    b,a = signal.cheby2(order, minimum_attenuation, Cutoff_Normalized, 'low', analog=False)  #a are the denominator coefficients.  
                                                                                                 #b are the numerator coefficients. 
    w, h = signal.freqz(b,a)    # w = 2pif, h = frequancy response in dB if 20 * np.log10(abs(h))

    
    plt.plot(0.5*Fs*w/np.pi, np.abs(h), 'b')  # w = 2pif with f = 0 to 1. The real frequency is Fs. 
    plt.ylabel('Magnitude')
    plt.xlabel('Frequency [Hz]')
    plt.title('low pass filter_Chebychev2')
    
    
    
```

```python

#Add 0 to can find x[k+2] and x[k+1] when k = 40. (Not 40 but 39 because we put k = 0 at the place of k = 1)
#Add 0 to can find x[k-2] and x[k-1] when k = 40. (Not 40 but 39 because we put k = 0 at the place of k = 1)

x = np.zeros(44)  #x = np.zeros(40)  #x = np.zeros(42) 
y1 = np.zeros(44)
y2 = np.zeros(44)
x[21] = 1  # x[19] = 1


s  = 1;
Ts = 1;

A = s*Ts; 
a = np.exp(-A);



for i in range (2,42):  #for i in range (0,40):
    k = 41 - i + 2 #k = 39 - i
    y1[k] = a * (A + 1) * x[k+1] - a**2 * x[k+2] + 2 * a * y1[k+1] - a**2*y1[k+2]
  
    
for k in range (2,42): #for i in range (0,40):
    y2[k] = x[k] + a*(A-1) * x[k-1] + 2*a*y2[k-1] - a**2*y2[k-2]
   



    
k = np.arange(44)
y = y1 + y2
plt.plot(k,x, 'o')
plt.plot(k,y,'o')
plt.title('in blue: x[k], in green: y[k]')




```
